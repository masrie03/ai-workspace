```python
# Import necessary libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
#in this code we are trying to predict the malaysia median house price
import kagglehub
from kagglehub import KaggleDatasetAdapter

df = kagglehub.load_dataset(
    KaggleDatasetAdapter.PANDAS,
    "lyhatt/house-prices-in-malaysia-2025",
    "malaysia_house_price_data_2025.csv"
)

df.head()
```

```python
df.info()
# linear regression needs numeric target(output) and numeric input features(predictors)
#currently we dont have any predictors. Based on the columns, "State" can be our predictor but not directly since the data type is object
##Hence we have to encode the "state" column later
```

```python
#set up our features(X) and target (y)
X = df[ ["State", "Tenure", "Type" , "Median_PSF", "Transactions"]]
Y = df["Median_Price"]
```

```python
# we are using one hot encoding, it creates binary column for each category (1 = TRUE, 0 = FALSE)
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

#define which columns are categorical and nominal
categorical_cols = ["State", "Tenure", "Type"]
numeric_cols = ["Median_PSF", "Transactions"]

preprocessor = ColumnTransformer(
    transformers=[
        ("cat", OneHotEncoder(drop="first", handle_unknown="ignore"), categorical_cols),
        ("num", "passthrough", numeric_cols)
    ]
)
#drop="first" to avoid duplicate
#"passthrough", numeric_cols keep numeric cols unchanged

X_encoded = preprocessor.fit_transform(X)
```
